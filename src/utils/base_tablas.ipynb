```python
'''
-- ==========================================================
-- ESTRUCTURA DE TABLAS
-- ==========================================================

CREATE TABLE ESPECIALIDAD (
    ID_Especialidad SERIAL PRIMARY KEY,
    Nombre_Especialidad VARCHAR(50) NOT NULL UNIQUE
);

CREATE TABLE CAMPUS (
    ID_Campus SERIAL PRIMARY KEY,
    Nombre_Campus VARCHAR(50) NOT NULL UNIQUE
);

CREATE TABLE PROFESORES (
    ID_Profesor SERIAL PRIMARY KEY,
    Nombre_Profesor VARCHAR(100) NOT NULL,
    Rol_Profesor VARCHAR(20) NOT NULL,
    Modalidad VARCHAR(20) NOT NULL,
    CHECK (Rol_Profesor IN ('TA', 'LI')),
    CHECK (Modalidad IN ('Online', 'Presencial'))
);

CREATE TABLE PROMOCION (
    ID_Promocion SERIAL PRIMARY KEY,
    Nombre_Promocion VARCHAR(50) NOT NULL,
    Fecha_Comienzo DATE NOT NULL,
    ID_Campus INT NOT NULL,
    ID_Especialidad INT NOT NULL,
    FOREIGN KEY (ID_Campus) REFERENCES CAMPUS(ID_Campus),
    FOREIGN KEY (ID_Especialidad) REFERENCES ESPECIALIDAD(ID_Especialidad)
);

CREATE TABLE ALUMNOS (
    ID_Alumno SERIAL PRIMARY KEY,
    Nombre_Alumno VARCHAR(100) NOT NULL,
    Email_Alumno VARCHAR(120) NOT NULL UNIQUE,
    ID_Promocion INT NOT NULL,
    FOREIGN KEY (ID_Promocion) REFERENCES PROMOCION(ID_Promocion)
);

CREATE TABLE PROYECTOS (
    ID_Proyecto SERIAL PRIMARY KEY,
    Nombre_Proyecto VARCHAR(100) NOT NULL UNIQUE,
    ID_Especialidad INT NOT NULL,
    FOREIGN KEY (ID_Especialidad) REFERENCES ESPECIALIDAD(ID_Especialidad)
);

CREATE TABLE PROFESOR_PROMOCION (
    ID_Profesor_Promocion SERIAL PRIMARY KEY,
    ID_Profesor INT NOT NULL,
    ID_Promocion INT NOT NULL,
    FOREIGN KEY (ID_Profesor) REFERENCES PROFESORES(ID_Profesor),
    FOREIGN KEY (ID_Promocion) REFERENCES PROMOCION(ID_Promocion),
    UNIQUE (ID_Profesor, ID_Promocion)
);

CREATE TABLE RESULTADOS (
    ID_Resultado SERIAL PRIMARY KEY,
    ID_Alumno INT NOT NULL,
    ID_Proyecto INT NOT NULL,
    Resultado_Final VARCHAR(20) NOT NULL,
    FOREIGN KEY (ID_Alumno) REFERENCES ALUMNOS(ID_Alumno),
    FOREIGN KEY (ID_Proyecto) REFERENCES PROYECTOS(ID_Proyecto),
    CHECK (Resultado_Final IN ('Apto', 'No Apto')),
    UNIQUE (ID_Alumno, ID_Proyecto)
);

-- ==========================================================
-- INSERTS MAESTROS
-- ==========================================================

INSERT INTO ESPECIALIDAD (Nombre_Especialidad) VALUES ('FS'), ('DS');
INSERT INTO CAMPUS (Nombre_Campus) VALUES ('Madrid'), ('Valencia');

COMMIT;
'''
```

4- Eliminado  CHECK (Nombre_Promocion IN ('Septiembre', 'Febrero')) de TABLE PROMOCION por tema de escalabilidad.

5- Usamos los Insert de Especialidad y Campus nos permite asegurar que los IDs de estas dos tablas sean siempre el mismo  y evitar futuros problemas.
