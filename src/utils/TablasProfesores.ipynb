```python
import pandas as pd
import sqlite3
```

```python
query = '''
CREATE TABLE IF NOT EXISTS Profesores (
    ID_Profesor INT,
    Nombre_profesor VARCHAR(100),
    Rol_Profesor VARCHAR(20),
    Modalidad VARCHAR(20),
    PRIMARY KEY (ID_Profesor)
)
'''

.execute(query)
```

```python
query = '''
CREATE TABLE IF NOT EXISTS Alumnos (
    ID_Alumno INT,
    Nombre_Alumno VARCHAR(100),
    Email_Alumno VARCHAR(100),
    ID_Promocion INT,
    PRIMARY KEY (ID_Alumno)
)
'''

.execute(query)
```

```python
query = '''
CREATE TABLE IF NOT EXISTS Promocion (
    ID_Promocion INT,
    Nombre_Promocion VARCHAR(100),
    Fecha_Comienzo INT,
    ID_Campus INT,
    ID_Especialidad INT,
    PRIMARY KEY (ID_Promocion)
)
'''

.execute(query)
```

```python
query = '''
CREATE TABLE IF NOT EXISTS Proyectos (
    ID_Proyecto INT,
    Nombre_Proyecto VARCHAR(100),
    ID_Especialidad INT,
    PRIMARY KEY (ID_Proyecto)
)
'''

.execute(query)
```

```python
query = '''
CREATE TABLE IF NOT EXISTS Campus (
    ID_Campus INT,
    Nombre_Campus VARCHAR(100),
    PRIMARY KEY (ID_Campus)
)
'''

.execute(query)
```

```python
query = '''
CREATE TABLE IF NOT EXISTS Resultados (
    ID_Resultado  INT,
    ID_Alumno INT,
    ID_Proyecto INT,
    Resultado_Final ,
    PRIMARY KEY (ID_Resultados)
)
'''

.execute(query)
```

```python
query = '''
CREATE TABLE IF NOT EXISTS Especialidad (
    ID_Especialidad INT,
    Nombre_Especialidad VARCHAR(100),
    PRIMARY KEY (ID_Especialidad)
)
'''

.execute(query)
```
